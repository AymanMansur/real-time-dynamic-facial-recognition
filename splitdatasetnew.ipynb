```python
import os
import shutil
import random
```

```python
# Path to your dataset directory containing folders 1 to 334
dataset_path = r"D:\thesis\FINAL ALL DATASET"

# Define paths for train, test, and validation sets
train_path = r"D:\thesis papers\train"
test_path = r"D:\thesis papers\test"
val_path = r"D:\thesis papers\val"
```

```python
# Create train, test, and validation directories if they don't exist
os.makedirs(train_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)
os.makedirs(val_path, exist_ok=True)

# Function to split and copy images to respective folders


def split_dataset(person_folder):
    # List all images for a person
    images = os.listdir(os.path.join(dataset_path, person_folder))

    # Shuffle the images randomly
    random.shuffle(images)

    # Calculate sizes for train, test, and validation sets
    total_images = len(images)
    print('ti',total_images)
    train_size = int(0.7 * total_images)
    print('train',train_size)
    test_size = int(0.2 * total_images)
    print('test',test_size)
    val_size = total_images - train_size - test_size
    print('val',val_size)

    # Split images into train, test, and validation sets
    train_images = images[:train_size]
    test_images = images[train_size:train_size + test_size]
    val_images = images[train_size + test_size:]

    # Copy images to respective train, test, and validation folders
    for img in train_images:
        src = os.path.join(dataset_path, person_folder, img)
        dst = os.path.join(train_path, person_folder, img)
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.copy(src, dst)

    for img in test_images:
        src = os.path.join(dataset_path, person_folder, img)
        dst = os.path.join(test_path, person_folder, img)
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.copy(src, dst)

    for img in val_images:
        src = os.path.join(dataset_path, person_folder, img)
        dst = os.path.join(val_path, person_folder, img)
        os.makedirs(os.path.dirname(dst), exist_ok=True)
        shutil.copy(src, dst)


# Iterate through each person folder and split the dataset
for i in range(1, 335):
    person_folder = str(i)
    split_dataset(person_folder)
```
